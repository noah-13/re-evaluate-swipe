```python
%%writefile script.py
import os
os.environ["MPLBACKEND"] = "Agg"
import matplotlib
matplotlib.use("Agg")

import pandas as pd
from questeval.questeval_metric import QuestEval

# Load CSV file
csv_filename = "input.csv"
df = pd.read_csv(csv_filename)

# Initialize QuestEval
questeval = QuestEval(no_cuda=False)

# Define model names
model_names = [
    'access',
    'keep_it_simple',
    'swipe_v5',
    'swipe_v5_clean',
    'gpt3',
    'bartl_wikilarge'
]

df_sample = df.copy()

# Get source text from 'r_content'
source_texts = df_sample['r_content'].tolist()
new_csv_filename = "questeval.csv"
df_sample.to_csv(new_csv_filename, index=False)
print(f"Sample CSV file has been updated and saved as: {new_csv_filename}")

hypotheses = df_sample['s_content'].tolist()
result = questeval.corpus_questeval(hypothesis=hypotheses, sources=source_texts)  # `source` not `sources`
df_sample.loc[:, "questeval_reference_free_reference"] = result["ex_level_scores"]
df_sample.to_csv(new_csv_filename, index=False)
print(f"Sample CSV file has been updated and saved as: {new_csv_filename}")

# Compute reference-free QuestEval scores for models
for name in model_names:
    hypotheses = df_sample['candidate_' + name].tolist()
    result = questeval.corpus_questeval(hypothesis=hypotheses, sources=source_texts)
    df_sample.loc[:, "questeval_reference_free_" + name] = result["ex_level_scores"]
    df_sample.to_csv(new_csv_filename, index=False)
    print(f"Sample CSV file has been updated and saved as: {new_csv_filename}")
# Compute reference-based QuestEval scores for models
reference_texts = df_sample['s_content'].tolist()
for name in model_names:
    hypotheses = df_sample['candidate_' + name].tolist()
    result = questeval.corpus_questeval(hypothesis=hypotheses, sources=reference_texts)
    df_sample.loc[:, "questeval_reference_based_" + name] = result["ex_level_scores"]
    df_sample.to_csv(new_csv_filename, index=False)
    print(f"Sample CSV file has been updated and saved as: {new_csv_filename}")

# Save the updated sample CSV file
df_sample.to_csv(new_csv_filename, index=False)

print(f"CSV file has been updated and saved as: {new_csv_filename}")



```
